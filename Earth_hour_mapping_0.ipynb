```
import json
from pathlib import Path

import folium
from folium import plugins
print('folium version is {}'.format(folium.__version__))
```

```
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```
# read csv
df_school = pd.read_csv('/content/drive/MyDrive/STAFF共有フォルダ：ことぶき協働スペース/7.小さなデジタル推進室/プログラミング Input/Programming/Colaboratory/Earth_Hour_2021/140007_public_facility_school.csv', encoding='SHIFT-JIS',engine='python')
```

```
# extract school of yokohama
df_school.head()
df_school_y = df_school[df_school['市町村名']=='横浜市']
df_school_y.head()
```

```
# draw basemap
map_center = [35.438393,139.638063] #協働スペースのlat long
earth_hour_map = folium.Map(location=map_center, tiles='cartodbpositron', zoom_start=15)

for i, dt in df_school_y.iterrows():
  folium.Marker(location=[dt['緯度'],dt['経度']], popup='{},{}'.format(dt['名称'],dt['創立年'])).add_to(earth_hour_map)
earth_hour_map
```

```
#read shape file using Geopandas
shdf = gpd.read_file('/content/drive/MyDrive/STAFF共有フォルダ：ことぶき協働スペース/7.小さなデジタル推進室/プログラミング Input/Programming/Colaboratory/Earth_Hour_2021/A27-10_14_GML/A27-10_14-g_SchoolDistrict.shp',encoding='SHIFT-JIS')
```

```
yokohama_esh = shdf[shdf['A27_006']=='横浜市立']
```

```

map_center = [35.438393,139.638063] #協働スペースのlat long
m_yokohama = folium.Map(location=map_center, tiles='cartodbpositron', zoom_start=14)
folium.Choropleth(geo_data=yokohama_esh.to_json()).add_to(m_yokohama)
m_yokohama
```
